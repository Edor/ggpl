```python
# Genero 100 punti nello spazio [0;100]^2 

listaPunti = []
for x in range (0,10,1):
    for y in range (0,10,1):
        listaPunti.append((randint(0,101),randint(0,101)))
listaPunti
```

```python
# Creo un modello unendo tutti e 100 i punti in un singolo poliedro. Con un po' di fortuna dovrebbe essere
# un poliedro di forma casuale inscritto nel quadrato di lato 100!

FlistaPunti = []
for x in range (0,len(listaPunti),1):
    FlistaPunti.append(x)

FlistaPunti2=[FlistaPunti]
model=listaPunti,FlistaPunti2
```

```python
# Visualizzo l'intero poliedro

VIEW(STRUCT(MKPOLS(model)))
```

```python
# Visualizzo solo il perimetro del poliedro

VIEW(SKEL_1(STRUCT(MKPOLS(model))))
```

```python
# Faccio una prova in 3D. Creo una nuova lista di punti.

NL = []
for x in range (0,5,1):
    for y in range (0,5,1):
        for z in range (0,5,1):
            NL.append((randint(0,101), randint(0,101), randint(30,71)))
NL
```
