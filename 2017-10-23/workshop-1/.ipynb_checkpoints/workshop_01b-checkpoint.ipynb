Facade script
=======

```python
from larlib import *
```

```python
# Definisco una nuova funzione chiamata facade che prende in input due interi positivi
def facade(n,m):
    """Return an HPC object representing a facade of n columns and m rows."""
    # Creo 4 nuovi array...
    # ... replicando n volte un array di base...
    xRithm = [5, -2, -5, -2]*n
    # ... e appendendo alla fine un 5...
    xRithm.append(5)
    # ... replicando m volte un diverso array ...
    yRithm = [7, -5, -2]*m
    # ... e appendendo alla fine un 7 ...
    yRithm.append(7)
    # Poi creo due copie di xRithm e yRithm le quali hanno tutti gli elementi inverititi di segno
    xVoid = [ -x for x in xRithm ]
    yVoid = [ -y for y in yRithm ]
    # Ritorno il prodotto cartesiano tra i 4 array opportunamente prodotto
    return STRUCT([PROD([QUOTE(xRithm),QUOTE(yRithm)]),PROD([QUOTE(xVoid),QUOTE(yRithm)]),PROD([QUOTE(xRithm),QUOTE(yVoid)]),SKEL_1(PROD([QUOTE(xVoid),QUOTE(yVoid)]))])
```

```python
# Please work...
VIEW(facade(6,4))
```
