```python
# Codice Python
Leg = CUBOID([0.1,0.1,0.7])
Plane = CUBOID([1,1,0.2])

# In questo modo ogni oggetto Leg o Plane è traslato rispetto alla traslazione totale
# di tutti gli oggetti che lo precedono
Table = STRUCT([Leg, T(1)(0.9), Leg, T(2)(0.9), Leg, T(1)(-0.9), Leg, T([2,3])([-0.9,0.7]), Plane])
VIEW(Table)
```

```python
# In questo modo la traslazione di ogni oggetto è assoluta rispetto al punto di origine (0,0,0)
Table1 = STRUCT([Leg, T(1)(0.9)(Leg), T([1,2])([0.9,0.9])(Leg), T(2)(0.9)(Leg), T(3)(0.7)(Plane)])
```
