```python
from larlib import *
import csv
```

```python
def pianta_HPC(pathToFile):
    with open(pathToFile, "rb") as f:
        reader = csv.reader(f, delimiter=",")
        
        keys = {}
        nodi = []
        archi = []
        i = 0
        
        for row in reader:
            row0 = float(row[0])
            row1 = float(row[1])
            row2 = float(row[2])
            row3 = float(row[3])
            if not (row0, row1) in keys.keys():
                keys[(row0, row1)] = i
                nodi.append((row0, row1))
                i = i+1
                
            if not (row2, row3) in keys.keys():
                keys[(row2, row3)] = i
                nodi.append((row2, row3))
                i = i+1
                
            archi.append((keys[(row0, row1)], (keys[(row2, row3)])))
    return (nodi, archi)
```

```python
E, EV = pianta_HPC("./villa_barbaro.lines")
```
