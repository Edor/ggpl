```python
from larlib import *
```

```python
verts=[[0,0],[10,0],[10,10],[5,13],[0,10]]
cells=[[1,2,3,4,5]]
pols=[]
```

```python
wall=MKPOL([verts,cells,pols])
wall
```

```python
VIEW(wall)
```

```python
door=CUBOID([3.5,7])
window=CUBOID([2,4])
house=STRUCT([wall,door,window])
```

```python
VIEW(SKEL_1(house))
```

```python
house=STRUCT([wall,T(1)(3)(door),T([1,2])([7,2])(window)])
```

```python
VIEW(SKEL_1(house))
```
