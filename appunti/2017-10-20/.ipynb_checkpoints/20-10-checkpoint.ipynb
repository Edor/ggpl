Appunti vari su traslazioni e altro
----------------------------------------

```python
from larlib import *
```

```python
verts=[[0,0],[10,0],[10,10],[5,13],[0,10]]
cells=[[1,2,3,4,5]]
pols=[]
```

```python
wall=MKPOL([verts,cells,pols])
wall
```

```python
VIEW(wall)
```

```python
door=CUBOID([3.5,7])
window=CUBOID([2,4])
house=STRUCT([wall,door,window])
```

```python
# Posiziono finestra e porta in luoghi pi adatti!
house=STRUCT([wall,T(1)(3)(door),T([1,2])([7,2])(window)])
```

```python
VIEW(SKEL_1(house))
```

```python
# Visualizzo la casa originale e una scalata di 1/2 su X e Y
VIEW(SKEL_1(STRUCT([house,S([1,2])([.5,.5])(house)])))
```

```python
# La funzione "scala" ribalta le immagini se viene passato un argomento negativo
VIEW(SKEL_1(STRUCT([house,S([1,2])([-1,-1])(house)])))
```

```python
VIEW(SKEL_1(STRUCT([house,R([1,2])(PI/3)(house)])))
```
